```python
%matplotlib inline
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
cases = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_US.csv")
death = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_US.csv")
```

```python
print(cases.head)
print(death.head)
```

```python
cases_CA = cases[cases["Province_State"] == "California"]
```

```python
cases_CA_indexed = cases_CA.set_index("Admin2")
cases_CA_T = cases_CA_indexed.T
```

```python
cases_CA_indexed
```

```python
cases_CA_T
```

```python
cases_clean = cases_CA_T.drop(["UID", "iso2", "iso3", "code3", "FIPS", "Lat", "Long_", "Combined_Key"])
```

```python
death_clean = death[death["Province_State"] == "California"].set_index("Admin2").T.drop(['UID','iso2','iso2','code3','FIPS','Lat','Long_','Combined_Key'])
```

```python
countties = ['Alameda',
             'San Francisco',
             'San Mateo',
             'Santa Clara']
```

```python
cases_clean_numeric = cases_clean.iloc[2:]
plot = cases_clean_numeric[countties].plot()
plot.set_title("COVID-19 cases in Bay area Counties")
```

```python
deaths_clean_numeric = cases_clean.iloc[2:]
plot = deaths_clean_numeric[countties].plot()
plot.set_title("COVID-19 deaths in Bay area Counties")
```

```python
cases_diff = cases_clean_numeric.diff().rolling(window=7).mean()
deaths_diff = deaths_clean_numeric.diff().rolling(window=7).mean()
```

```python
pop = pd.read_csv("https://gist.githubusercontent.com/NillsF/7923a8c7f27ca98ec75b7e1529f259bb/raw/3bedefbe2e242addba3fb47cbcd239fbed16cd54/california.csv")
```

```python
pop["CTYNAME"] = pop['CTYNAME'].str.replace("country" , "")
```

```python
pop
```

```python
pop2 = pop.drop("GrowthRate", axis=1)
pop2["CTYNAME"] = pop2["CTYNAME"].str.replace(" County", "")
pop2 = pop2.set_index('CTYNAME')
```

```python
pop2
```

```python
death_clean_numeric = death_clean.iloc[5:]
for col in death_clean_numeric.columns:
  death_clean_numeric[col] = pd.to_numeric(death_clean_numeric[col], errors='coerce')

cases_pm = cases_clean_numeric.copy()
deaths_pm = death_clean_numeric.copy()

# Remove ' County' from the column names to match the population data
cases_pm.columns = cases_pm.columns.str.replace(" County", "")
deaths_pm.columns = deaths_pm.columns.str.replace(" County", "")

for c in pop2.index.tolist():
    cases_pm[c] = cases_pm[c] / pop2.loc[c, :]['Pop'] * 100000
    deaths_pm[c] = deaths_pm[c] / pop2.loc[c, :]['Pop'] * 100000
```

```python
cases_pm_diff = cases_pm.diff().rolling(window=7).mean()
deaths_pm_diff = deaths_pm.diff().rolling(window=7).mean()
```

```python
plot = cases_diff[countties].plot()
plot.set_title("7 days moving avg of new COVID-19 cases ")
```

```python
plot = cases_pm_diff[countties].plot()
plot.set_title("7 days moving avg of new COVID-19 cases ")
```

```python
plot = cases_pm.sort_values(axis=1,by='7/20/20',ascending=False).iloc[:,:10].plot(figsize=(20,10))
plot.set_title("Top 10 countries by COVID-19 cases per million inhabitants")
```

```python
plot = cases_pm_diff.sort_values(axis=1,by='7/20/20',ascending=False).iloc[:,:10].plot(figsize=(20,10))
plot.set_title("Top 10 countries by 7 days rolling avg COVID-19 case increases  per million inhabitants")
```

```python
plot = deaths_pm.sort_values(axis=1,by='7/20/20',ascending=False).iloc[:,:10].plot(figsize=(20,10))
plot.set_title("Top 10 countries by 7 days rolling avg COVID-19 case increases  per million inhabitants")
```

```python
plot = deaths_pm_diff.sort_values(axis=1,by='7/20/20',ascending=False).iloc[:,:10].plot(figsize=(20,10))
plot.set_title("Top 10 countries by 7 days rolling avg COVID-19 case increases  per million inhabitants")
```
